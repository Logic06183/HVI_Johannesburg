## GHCN

The Global Historical Climatology Network daily (GHCNd) is an integrated database of daily climate summaries from land surface stations across the globe. GHCNd is made up of daily climate records from numerous sources that have been integrated and subjected to a common suite of quality assurance reviews.

```python
metadata_gdf = gpd.read_file( "/terra/projects/heat_center/data/climate/observed/GHCN/GHCN_sub_saharan_metadata.geojson")
metadata_gdf
```

```python
ds_station = xr.open_zarr('/terra/projects/heat_center/data/climate/observed/GHCN/ghcnd_archive.zarr')
ds_station
```

If we want to select a single station closest to a given point

```python
from geopy.distance import great_circle

target_lat, target_lon = -26.134734, 28.238075 # OR Tambo International Airport 
target_point = Point(target_lon, target_lat)

# Calculate distances using the Haversine formula
metadata_gdf['Distance'] = metadata_gdf.apply(lambda row: great_circle((row['geometry'].y, row['geometry'].x), (target_lat, target_lon)).kilometers, axis=1)
closest_point = metadata_gdf.loc[metadata_gdf['Distance'].idxmin()]
closest_point
```

Only 1 km away

```python
ds_station.sel(ghcn_id = closest_point.station_id).tmax.plot()
```

Let's plot the locations of these stations 

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

for idx, row in metadata_gdf.iterrows():
    ax.plot(row.geometry.x, row.geometry.y, marker='x', color='red', markersize=2, transform=ccrs.PlateCarree())

ax.coastlines()
plt.show()
```

less good when considering those with a recent 20 year time span

```python
subset = metadata_gdf[metadata_gdf.start_year < 2000]
subset = subset[subset.end_year > 2020]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

for idx, row in subset.iterrows():
    ax.plot(row.geometry.x, row.geometry.y, marker='x', color='red', markersize=2, transform=ccrs.PlateCarree())

ax.coastlines()
plt.show()
```

```python
metadata_gdf.crs = 'EPSG:4326'
polygon.crs = 'EPSG:4326'


# Perform a spatial join to subset the GeoDataFrame to include only points within the polygon
subset_gdf = gpd.sjoin(metadata_gdf, polygon, how="inner", predicate='within')
subset_gdf
```
